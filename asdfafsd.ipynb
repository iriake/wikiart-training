```python
import csv
from collections import Counter
from pathlib import Path

csv_path = Path("/kaggle/input/wikiart/classes.csv")

if not csv_path.exists():
    csv_path = Path("CSVs/classes.csv")

rows = []

with csv_path.open(newline="", encoding="utf-8") as csv_file:
    reader = csv.DictReader(csv_file)

    for row in reader:
        row["class_name"] = row["filename"].split("/", 1)[0]
        rows.append(row)

class_counts = Counter(row["class_name"] for row in rows)
subset_counts = Counter(row["subset"] for row in rows)

print(f"CSV: {csv_path}")
print(f"Rows: {len(rows)}")
print(f"Classes: {len(class_counts)}")

print("\nSubset counts:")
for subset, count in sorted(subset_counts.items()):
    print(f"{subset}: {count}")

print("\nClass counts:")
for class_name, count in sorted(class_counts.items()):
    print(f"{class_name}: {count}")
```
